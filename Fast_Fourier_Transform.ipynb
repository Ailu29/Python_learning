```python
import numpy as np
import matplotlib.pyplot as plt

# 
fs = 1000  # 
t = np.linspace(0, 1, fs, endpoint=False)  # 
f = 5  # 5Hz
signal = np.sin(2 * np.pi * f * t)

# 
fft_result = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(signal), 1/fs)

# FFT
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.plot(t, signal)
plt.title('Original Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(122)
plt.stem(frequencies, np.abs(fft_result), 'b', markerfmt=" ", basefmt="-b")
plt.title('FFT of the signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.xlim(0, fs/2)  # 
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
Fs = 1000  # 
T = 1/Fs   # 
L = 1000   # 
t = np.arange(0, L) * T  # 

# 
f1 = 50  # 
f2 = 120 # 
signal = 0.7*np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2*t)

# FFT
fft_result = np.fft.fft(signal)

# 
frequencies = np.fft.fftfreq(L, T)

# 
magnitude_spectrum = np.abs(fft_result)

# 
phase_spectrum = np.angle(fft_result)

# 
real_part = np.real(fft_result)
imaginary_part = np.imag(fft_result)

# 
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(frequencies[:L//2], magnitude_spectrum[:L//2])
plt.title('Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.subplot(3, 1, 2)
plt.plot(frequencies[:L//2], phase_spectrum[:L//2])
plt.title('Phase Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (radians)')

plt.subplot(3, 1, 3)
plt.plot(frequencies[:L//2], real_part[:L//2], label='Real Part')
plt.plot(frequencies[:L//2], imaginary_part[:L//2], label='Imaginary Part')
plt.title('Real and Imaginary Parts')
plt.xlabel('Frequency (Hz)')
plt.legend()

plt.tight_layout()
plt.show()
```
